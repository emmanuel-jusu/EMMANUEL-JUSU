[Problem 1] Creating random numbers

```python
import numpy as np
mean = [-3, 0]
cov = [[1.0, 0.8], [0.8, 1.0]]
seed = np. random.seed(0)
rand_number = np.random.multivariate_normal(mean, cov, 500)
print(rand_number)
rand_number.shape
```

[problem 2] Visualization by scatter plot

```python
from matplotlib import pyplot as plt
%matplotlib inline
plt.title("scatter plot")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(rand_number[:,0], rand_number[:,1])
```

[Problem 3] Visualization by histogram

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.xlabel('x1')
plt.ylabel('frequency')
plt.title('histogram of x1')
plt.hist(rand_number[:,0])
plt.xlim(left=-5)
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.xlabel('x2')
plt.ylabel('frequency')
plt.title('histogram of x2')
plt.hist(rand_number[:,1])
plt.xlim(right=2)
plt.show()
```

[Problem 5] Data combination

```python
#Using concatenate
data_combine = np.concatenate((rand_number, rand_number1),axis=0)
print(data_combine)
data_combine.shape
```

```python
## Using vstack
combined_data = np.vstack((rand_number, rand_number1))
combined_data.shape
```
